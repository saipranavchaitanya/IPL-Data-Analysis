```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
```

```python
matches = pd.read_csv('matches .csv')
deliveries = pd.read_csv('deliveries.csv')
```

```python
matches.head()
matches.info()
```

```python
deliveries.head()
deliveries.info()
```

```python
matches = matches[matches['result'] == 'normal']
```

```python
team_map = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Rising Pune Supergiants': 'Rising Pune Supergiant'
}

for col in ['team1','team2','winner']:
    matches[col] = matches[col].replace(team_map)
```

```python
matches_per_season = matches['season'].value_counts().sort_index()
matches_per_season
```

```python
matches_per_season.plot(kind='bar')
plt.title('Matches per Season')
plt.show()
```

```python
team_wins = matches['winner'].value_counts()
team_wins.head(10)
```

```python
toss_win_match_win = matches[matches['toss_winner'] == matches['winner']]
percentage = round((len(toss_win_match_win)/len(matches))*100,2)
percentage
```

```python
top_batsmen = deliveries.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False)
top_batsmen.head(10)
```

```python
balls = deliveries.groupby('batter')['ball'].count()
runs = deliveries.groupby('batter')['batsman_runs'].sum()

strike_rate = (runs/balls)*100
strike_rate = strike_rate[balls > 500].sort_values(ascending=False)
strike_rate.head(10)
```

```python
wickets = deliveries[deliveries['dismissal_kind'].notnull()]
top_bowlers = wickets['bowler'].value_counts()
top_bowlers.head(10)
```

```python
venue_matches = matches['venue'].value_counts()
venue_matches.head(10)
```

```python
team_wins.to_csv('team_wins.csv')
top_batsmen.to_csv('top_batsmen.csv')
strike_rate.to_csv('strike_rate.csv')
top_bowlers.to_csv('top_bowlers.csv')
venue_matches.to_csv('venue_matches.csv')
matches_per_season.to_csv('matches_per_season.csv')
```
